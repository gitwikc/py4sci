```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt, style
```

```python
style.use("ggplot")z
```

```python
def plot_wave(A=1, lambd=1, omega_t=0):
    x = np.linspace(-2 * lambd, 0, 100)
    y1 = A * np.cos(omega_t - (2 * np.pi / lambd) * x)
    y2 = -A * np.cos(omega_t + (2 * np.pi / lambd) * x)
    y = y1 + y2
    df = pd.DataFrame(
        data=dict(
            x=x,
            y1=y1,
            y2=y2,
            y=y
        ),
        columns=["x", "y1", "y2", "y"],
    )
    df.plot(x="x")
    plt.hlines(y=0, xmin=-2 * lambd, xmax=0, colors="k", linestyles="dotted")
```

```python
plot_wave(omega_t=np.pi / 4)
```

```python
plot_wave(omega_t=np.pi / 2)
```
